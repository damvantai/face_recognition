## Extract image from folder image

```python
import glob
import dlib
import cv2
import matplotlib.pyplot as plt
```

```python
image_list_file = []
for filename in glob.glob("../data/extract_face/anh dung/" + "*.jpg"):
    image_list_file.append(filename)
```

```python
detector = dlib.get_frontal_face_detector()
count = 1
```

```python
for filename in image_list_file:
    image = cv2.imread(filename, 1)
    detected_face = detector(image, 1)
    
    for i, d in enumerate(detected_face):
        x1, y1, x2, y2= d.left(), d.top(), d.right() + 1, d.bottom() + 1
#         cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
        
        save_image = image[y1-10: y2+10, x1-10: x2+10]
        name = "../data/extract_face/anh dung/face/Doan Van Dung." + str(count) + ".jpg"
        cv2.imwrite(name, save_image)
        count += 1
```
